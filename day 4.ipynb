# education column

```python
df['education'].value_counts()
```

```python
df['education'] = df['education'].replace({
    "Preschool":"Secondary",
    "1st-4th":"Secondary",
    "5th-6th":"Secondary",
    "7th-8th":"Secondary",
    "9th":"Secondary",
    "10th":"Secondary",
    "11th":"High School",
    "12th":"High School",
    "HS-grad":"High School",
    "Some-college":"Associate",
    "Assoc-voc":"Associate",
    "Assoc-acdm":"Associate",
    "Prof-school":"Associate"
}
)
```

```python
df['education'].value_counts()
```

```python
education = df['education'].value_counts()
sns.countplot(data= df, x='education', hue='income')
plt.title('Education level of account holders')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(rotation=40)
plt.yticks(rotation=40)
plt.show()
```

```python
size_mapping = {'Secondary':0, 'High School':1, 'Associate':2, 'Bachelors':3, 'Masters':4,  'Doctorate':5}
df['education'] = df['education'].replace(size_mapping)
df
```
