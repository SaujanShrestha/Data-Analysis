```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('weather_data.csv')
df
```

```python
#1 Simulate temperature data for 7 days in Celsius (morning, afternoon, and evening temperatures as a 2D array).
import numpy as np
morning_temp = df['Morning Temp (C)'].head(50)
afternoon_temp = df['Afternoon Temp (C)'].head(50)
evening_temp = df['Evening Temp (C)'].head(50)
temperature_data = np.column_stack((morning_temp, afternoon_temp, evening_temp))
print("7-Day Temperature Data (°C):",temperature_data)
```

```python
#2 Calculate:
    #a)The daily average temperature.
    #b)The day with the highest and lowest average temperature.
    #c)Convert all temperatures to Fahrenheit and display both Celsius and Fahrenheit values.

#a)
daily_average = np.mean(temperature_data,axis=1)
print("7-Day Temperature Data (°C):",daily_average)
```

```python
#b)
hottest_day = np.argmax(daily_average)
coldest_day = np.argmin(daily_average)
print(f"Hottest day (Day {hottest_day + 1}): {(daily_average[hottest_day]):.2f}°C")
print(f"Coldest day (Day {coldest_day + 1}): {(daily_average[coldest_day]):.2f}°C")
```

```python
#c)
celsius =  temperature_data
fahrenheit = celsius * 9/5 + 32
result = np.column_stack((celsius, fahrenheit))
print("Celsius to Fahrenheit Conversion:")
print("Celsius | Fahrenheit")
print(result)
```

```python
#3) Plot the temperature trends using Matplotlib (optional extension).
plt.figure(figsize=(10, 5))
plt.plot(result, marker='o', linestyle='-', color='skyblue')
plt.title('Monthly Temperature Trend ')
plt.xlabel('Month')
plt.ylabel('Temperature (°F)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4) Find Missing Numbers:
    #a)Given an array of numbers from 1 to 100 with some missing values, identify the missing numbers.
    #b)Input Format: A 1D array of numbers.
    #c)Output Format: A 1D array of missing numbers.


def find_missing_numbers(arr):
    arr = np.array(arr)
    full_set = np.arange(1, 101)
    missing = np.setdiff1d(full_set, arr)
    return missing
input_array = np.array([1, 2, 4, 5, 55, 62, 68, 73, 81, 100])
missing = find_missing_numbers(input_array)
print("Missing numbers:", missing)
```

```python
#5) Count Frequency:
    #a)Count the frequency of each unique value in a given 1D array.
    #b)Input Format: A 1D array with repeated values.
    #c)Output Format: A dictionary with values as keys and their frequencies as values.

#a)
def count_frequencies(arr):
    arr = np.array(arr)
    unique, counts = np.unique(arr, return_counts=True)
    frequency_dict = dict(zip(unique, counts))
    return frequency_dict
#b)
input_array = np.array([1, 2, 2, 3, 1, 4, 2])
freq = count_frequencies(input_array)
#c)
print("Frequencies:", freq)
```

```python
#6) Matrix Comparison:
    #a)Compare two matrices and identify which elements differ.
    #b)Input Format: Two 2D arrays of the same shape.
    #c)Output Format: Indices of differing elements.

#a)
def compare_matrix(mat1, mat2):
    mat1 = np.array(mat1)
    mat2 = np.array(mat2)
    diff_indices = np.argwhere(mat1 != mat2)
    return diff_indices
#b)
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[1, 9, 3],
              [0, 5, 7]])
#c)
print(compare_matrix(A, B))
```

```python
#7) Filter Rows by Condition:
    #a)Filter rows from a 2D array where the sum of elements in the row is greater than a given value.
    #b)Input Format: A 2D array and a threshold value.
    #c)Output Format: Filtered rows as a 2D array.

def filter_rows_by_sum(arr, threshold):
    return [row for row in arr if sum(row) > threshold]
array = [[1, 2, 3],
    [4, 5, 6],
    [1, 1, 1],
    [10, 10, 10]]
threshold_value = 10
filtered_array = filter_rows_by_sum(array, threshold_value)
print("Filtered Rows:")
print(filtered_array)
```

```python
#8) Random Sampling
    #a)Generate a random sample of 10 numbers from a given array without replacement.
    #b)Input Format: A 1D array with at least 10 elements.
    #c)Output Format: A 1D array with 10 random samples.

arr = np.array([5, 12, 7, 9, 14, 3, 8, 10, 6, 1, 4, 2, 11])
sample = np.random.choice(arr, size=10)
print("Random Sample:", sample)

```

```python
import pandas as pd
df = pd.read_csv("sales_data.csv")
df
```

```python
#1) Total Revenue: Calculate the total revenue from all sales.
df['Price'] = df['Quantity'] * df['Price']
total_revenue = df['Price'].sum()
print("Total Revenue:", total_revenue)
```

```python
#2) Best-Selling Products: Identify the top 5 products by total quantity sold.
product_sales = df.groupby('Product Name')['Quantity'].sum()
top_5_products = product_sales.sort_values(ascending=False).head(5)
top_5_products
```

```python
#3) Sales by Category: Group sales data by Category and calculate: 
    #a)Total sales. 
    #b)Average sales per product.
#a)
total_sales = df.groupby('Category')['Price'].sum()
#b)
average_sales = df.groupby('Category')['Price'].mean()
Final = pd.DataFrame({'Total Sales': total_sales, 'Average Sales per Product': average_sales})
Final
```

```python
#4) Regional Performance: Analyze sales by Region and identify the region with the highest revenue.
region_sales = df.groupby('Region')['Price'].sum()
top_region = region_sales.idxmax()
high_revenue = region_sales.max()
region_sales
print(f"Top Region: {top_region} with sales of ${high_revenue}")
```

```python
#5) Monthly Trends:
    #a)Extract the month from Order Date.
    #b)Group data by month to calculate total revenue for each month.
#a)
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month_name()

#b)
monthly_revenue = df.groupby('Month')['Price'].sum()
month_order = ['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']
pivot_table = pd.pivot_table(
    df,
    values='Price',       
    index='Month',        
    aggfunc='sum'         
)
monthly_revenue = pivot_table.reindex(month_order)
monthly_revenue
```

```python

#6) Customer Analysis:Identify the top 5 customers based on total spending.

customer_spending = df.groupby('Customer ID')['Price'].sum()
top_customers = customer_spending.sort_values(ascending=False).head(5)
pivot_table = top_customers.reset_index().pivot_table(
    index='Customer ID',
    values='Price',
    aggfunc='sum'
)
pivot_table
```

```python
#Task 25: Hourly Sales Trend
    #Objective: Visualize hourly sales data for a single day using a line chart.
    #Input Format:
    #A dataset CSV file (task_25_data.csv) with columns:
    #Hour (int): Hour of the day (0–23).
    #Total Revenue (float): Revenue during that hour.

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("task_25_data.csv")
plt.figure(figsize=(10, 6))
plt.plot(df['Hour'], df['Total Revenue'], marker='o', linestyle='-', color='blue')
plt.title('Hourly Sales Trend')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Revenue')
plt.xticks(range(0, 24))
plt.grid()
plt.show()
```

```python
#Task 26: Average Revenue by Product Category
    #Objective: Create a bar chart showing the average revenue by product category.
    #Input Format:
    #A dataset CSV file (task_26_data.csv)with columns:
    #Category (string): Product category.
    #Revenue (float): Revenue generated.

df = pd.read_csv("task_26_data.csv")
avg_revenue = df.groupby('Category')['Revenue'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(avg_revenue.index, avg_revenue.values, color='blue')
plt.title('Average Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
#Task 27: Sales Heatmap by Day and Region
    #Objective: Create a heatmap to show sales for each region on different days of the week.
    #Input Format:
    #A dataset CSV file (task_27_data.csv) with columns:
    #Day (string): Day of the week.
    #Region (string): Region name.
    #Total Sales (float): Sales value.

import seaborn as sns
df = pd.read_csv('task_27_data.csv')
pivot = df.pivot_table(index='Day', columns='Region', values='Total Sales', aggfunc='sum')
order = ['Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
pivot = pivot.reindex(order)
sns.heatmap(pivot, annot=True, cmap='Blues')
plt.title('Sales Heatmap by Day and Region')
plt.show()
```

```python
#Task 28: Product Demand Trends
    #Objective: Use a multi-line plot to show demand trends for multiple products over time.
    #Input Format:
    #A dataset CSV file (task_28_data.csv)with columns:
    #Date (datetime): Date of sale.
    #Product (string): Product name.
    #Quantity Sold (int): Units sold.


df = pd.read_csv('task_28_data.csv', parse_dates=['Date'])
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Quantity Sold', hue='Product', marker='o')
plt.title('Product Demand Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.legend(title='Product')
plt.xticks(rotation=45)
plt.show()
```

```python
#Task 29: Sales Growth Comparison Across Regions
    #Objective: Compare monthly sales growth across different regions using a grouped bar chart.
    #Input Format:
    #A dataset CSV file (task_29_data.csv) with columns:
    #Month (string): Name of the month.
    #Region (string): Region name.
    #Total Revenue (float): Revenue value.

df = pd.read_csv('task_29_data.csv')
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Month', y='Total Revenue', hue='Region')
plt.title('Monthly Sales Growth Comparison Across Regions')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
#Task 30: Outlier Analysis in Sales Data
    #Objective: Use a box plot to identify outliers in sales data.
    #Input Format:
    #A CSV file (task_30_data.csv) column with Total Sales (float).

df = pd.read_csv('task_30_data.csv')
plt.figure(figsize=(7, 7))
sns.boxplot(y=df['Total Sales'])
plt.title('Outlier Analysis in Sales Data')
plt.ylabel('Total Sales')
plt.show()
```
